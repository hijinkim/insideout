```python
import json
import pandas as pd
```

```python
import glob

path = 'json/*'
json_list = glob.glob(path) # get json file name list
```

```python
result = []

for j_file in json_list:
  with open(j_file) as f:
    json_data = json.load(f)
    label = list(json_data['flags'].values())
    img_path = json_data['imagePath']
    img_path = img_path.replace('.png', '')
    label.append(img_path)

  result.append(label)
print(result)
    
```

```python
col = json_data['flags'].keys()
col = list(col)
col.append('img_path')
col
```

```python
df = pd.DataFrame(data=result, columns=col)
```
