```python
import glob

path = 'json/*'
file_list = glob.glob(path) # get json file name list

len(file_list)
```

```python
replaced_list = []
for f in file_list:
  r = f.replace('json/','')
  r = r.replace('.json', '')
  replaced_list.append(r)

len(replaced_list)
```

```python
import glob

path = 'img/*'
img_list = glob.glob(path) # get image file name list

len(img_list)
```

```python
cnt = 0
img_replaced_list = []

for f in img_list:
  r = f.replace('img/', '')
  r = r.replace('.png', '')
  img_replaced_list.append(r)

len(img_replaced_list)
    
# os.remove('5.PNG')
```

```python
result = []
for i in img_replaced_list:
  if i not in replaced_list:
    result.append(i)

len(result)
```

```python
import os

for i in result:
  name = 'img/' + i + '.png'
  print(name)
  os.remove(name)
```
