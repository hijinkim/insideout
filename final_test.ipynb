```python
loaded_model = load_model('best_model_0527_1.h5')
```

```python
def predict_img(filename):
  img = image.load_img(filename, target_size=(256, 256))

  img = image.img_to_array(img)

  img = img/255.

  classes = np.array(['root', 'branch', 'fruit', 'leaves', 'knot'])
  pred = loaded_model.predict(img.reshape(1, 256, 256, 3))

  pred = (pred > 0.5).astype(int)

  result = list(pred[0])

  return result, img
```

```python
test = pd.read_excel('test_data.xlsx')
test = test.drop('Unnamed: 0', axis=1)
test
```

```python
y_test = np.array(test.drop('img_path', axis=1))
y_test
```

```python
cnt = 0
X_test = []

y_pred_total = []
y_test_total = []

for i in range(len(test['img_path'])):
  y_pred, img = predict_img(test['img_path'][i])
  X_test.append(img)
  y_pred_total += y_pred
  y_test_total += list(y_test[i])
  print(i+1)
  print(y_pred)
  print(list(y_test[i]), end='\n\n')

  if (y_pred == y_test[i]).all():
    cnt += 1

print(f'{cnt} / 24')
```

```python
cnt2 = 0
for i in range(len(y_test_total)):
  if y_pred_total[i] == y_test_total[i]:
    cnt2 += 1

print(f'{cnt2} / {len(y_test_total)}')
```

```python
X_test = np.array(X_test)
X_test.shape
```

```python
loss, accuracy = loaded_model.evaluate(X_test, y_test)

print('Loss = {:.5f}'.format(loss))
print('Accuracy = {:.5f}'.format(accuracy))
```

```python
y_pred = loaded_model.predict(X_test)
y_pred = (y_pred > 0.5).astype(int)

multilabel_confusion_matrix(y_test, y_pred)
```

```python
from sklearn import metrics

print(metrics.classification_report(y_test, y_pred))
```
